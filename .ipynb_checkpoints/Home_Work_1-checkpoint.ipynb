#### Сгенерируем обучающую выборку

```python
len_1 = 150
len_2 = 150

set_1 = scipy.stats.multivariate_normal.rvs(size=len_1, mean=[1, 2], cov=[[2.5, 1.5], [1.5, 2.5]])
set_2 = scipy.stats.multivariate_normal.rvs(size=len_2, mean=[6, 5], cov=[[2, -0.5], [-0.5, 2]])
```

```python
X_train = np.vstack([set_1, set_2])
y_train = np.hstack([np.zeros(len_1), np.ones(len_2)])
print (X_train.shape, y_train.shape)
```

```python
plt.figure(figsize=(15, 10))
plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train, alpha=0.5, cmap='winter')
plt.show()
```

#### Сгенерируем тестовую выборку

```python
set_1 = scipy.stats.multivariate_normal.rvs(size=2*len_1, mean=[1, 2], cov=[[2.5, 1.5], [1.5, 2.5]])
set_2 = scipy.stats.multivariate_normal.rvs(size=2*len_2, mean=[6, 5], cov=[[2, -0.5], [-0.5, 2]])
```

```python
X_test = np.vstack([set_1, set_2])
y_test = np.hstack([np.zeros(2 * len_1), np.ones(2*len_2)])
print (X_train.shape, y_train.shape)
```

```python
plt.figure(figsize=(15, 10))
plt.scatter(X_test[:, 0], X_test[:, 1], c=y_test, alpha=0.5, cmap='winter')
plt.xlim((min(X_test[:, 0]) - 0.5, max(X_test[:, 0]) + 0.5))
plt.ylim((min(X_test[:, 1]) - 0.5, max(X_test[:, 1]) + 0.5))
plt.show()
```

#### Обучим модель

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import zero_one_loss
from sklearn.metrics import accuracy_score
```

```python
model = KNeighborsClassifier(n_neighbors=3, algorithm='brute')
model.fit(X_train, y_train)
print(model)
```

#### Изобразим результаты работы классификатора

```python
def generate_grid(train_sample, border=1, step=0.05):
    return np.meshgrid(np.arange(min(train_sample[:, 0]) - border, max(train_sample[:, 1]) + border, step), 
                       np.arange(min(train_sample[:, 1]) - border, max(train_sample[:, 1]) + border, step))
```

```python
def create_picture(X_train, y_train, model, border=1, step=0.05, figsize=(18, 5), cmap='summer', alpha=1):
    
    # Создание сетки
    grid = generate_grid(X_train, border, step)
    grid_ravel = np.c_[grid[0].ravel(), grid[1].ravel(0)]  # Выворачивание сетки
    
    # Предсказание значений для сетки
    grid_predicted_ravel = model.predict_proba(grid_ravel)[:, 1]  # Берем вероятности для первого класса
    grid_predicted = grid_predicted_ravel.reshape(grid[0].shape)  # Подгоняем размер
    
    
    # Построение фигуры
    plt.figure(figsize=figsize)
    
    plt.subplot(1, 2, 1)
    plt.pcolormesh(grid[0], grid[1], grid_predicted > 0.5, cmap=cmap)
    plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train, alpha=alpha, cmap=cmap)
    plt.xlim((min(grid_ravel[:, 0]), max(grid_ravel[:, 0])))
    plt.ylim((min(grid_ravel[:, 1]), max(grid_ravel[:, 1])))
    plt.title(u'Classification {}-NN'.format(model.get_params()['n_neighbors']))
    
    plt.show()
    
    
    # Удаление лишних данных
    del(grid)
    del(grid_ravel)
    del(grid_predicted_ravel)
    del(grid_predicted)
```

```python
for k in [1, 3, 5, 10, 15, 20, 50, 100]:
    create_picture(X_train, y_train, KNeighborsClassifier(n_neighbors=k).fit(X_train, y_train))
```

#### Cross validation

```python
from sklearn.model_selection import GridSearchCV
```

```python
params = {'n_neighbors': list(range(1, 201))}
grid_searcher = GridSearchCV(KNeighborsClassifier(), params, cv=5, n_jobs=4)
grid_searcher.fit(X_train, y_train)
```

```python
means = np.array(grid_searcher.cv_results_['mean_test_score'])
```

```python
plt.figure(figsize=(15, 6))
plt.plot(range(1, 201), means)
plt.xlabel('Number of nearest neighbous')
plt.ylabel('Accuracy')
plt.ylim(0.90, 1)
plt.show()
```

```python
print("Оптимальное количество соседей: ", grid_searcher.best_params_['n_neighbors'])
```

```python
from sklearn import datasets, linear_model
from pandas import DataFrame
```

```python
digits = datasets.load_digits()
breast_cancer = datasets.load_breast_cancer()
```

```python
digits_frame = DataFrame(digits.data)
digits_frame['target'] = digits.target
```

```python
digits_frame.head()
```

```python
y_digits = digits_frame['target'].as_matrix()
X_digits = digits_frame.drop('target', axis=1).as_matrix()
```

```python
breast_cancer_frame = DataFrame(breast_cancer.data)
breast_cancer_frame.columns = breast_cancer.feature_names
breast_cancer_frame['target'] = breast_cancer.target
```

```python
breast_cancer_frame.head()
```

```python
y_cancer = breast_cancer_frame['target'].as_matrix()
X_cancer = breast_cancer_frame.drop('target', axis=1).as_matrix()
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.naive_bayes import BernoulliNB
from sklearn.naive_bayes import MultinomialNB
from sklearn.naive_bayes import GaussianNB
```

```python
BernNB = BernoulliNB()
MultyNB = MultinomialNB()
GaussNB = GaussianNB()
```

```python
print("Результат работы BernolliNB на датасете Digits:",cross_val_score(BernNB, X_digits, y_digits).mean())
print("Результат работы BernolliNB на датасете Breast Cancer:",cross_val_score(BernNB, X_cancer, y_cancer).mean())
print()
print("Результат работы MultinomialNB на датасете Digits:",cross_val_score(MultyNB, X_digits, y_digits).mean())
print("Результат работы MultinomialNB на датасете Breast Cancer:",cross_val_score(MultyNB, X_cancer, y_cancer).mean())
print()
print("Результат работы GaussianNB на датасете Digits:",cross_val_score(GaussNB, X_digits, y_digits).mean())
print("Результат работы GaussianNB на датасете Breast Cancer:",cross_val_score(GaussNB, X_cancer, y_cancer).mean())
```

Совпало ли это с ожиданиями? Да. 

Было вполне ожидаемо, что целые неотрицательные числа не очень хорошо приближаются нормальным распределением, а вот мультиномиальное - это как раз то, что нужно. Казалось, что биномимальное тоже должно неплохо отработать, но оно отработало сильно хуже мультиномиального (видимо, потому что признаки бинаризуются и теряется какая-то часть из них или связь между ними). 

По поводу вещественных признаков - понятно, что нормальное должно хорошо их приближать, потому что эти величины в среднем примерно одинаковые, и у них есть какя-то диспресия (это видно из названий признаков), а вот мультиномиальное и биномиальное вряд ли дадут хороший ответ, потому что слишком много возможных различных значений для каждого признака.

```python
print("Максимальное качество работы на датасете Digits:", max(cross_val_score(BernNB, X_digits, y_digits).max(), 
                                                             cross_val_score(MultyNB, X_digits, y_digits).max(),
                                                             cross_val_score(GaussNB, X_digits, y_digits).max()))
```

```python
print("Максимальное качество работы на датасете Breast Cancer:", max(cross_val_score(BernNB, X_cancer, y_cancer).max(), \
                                                             cross_val_score(MultyNB, X_cancer, y_cancer).max(),
                                                             cross_val_score(GaussNB, X_cancer, y_cancer).max()))
```

(a) На вещественных признаках лучше всего сработал наивный байесовский классификатор с распределением Бернулли.

(b) На вещественных признаках лучше всего сработал наивный байесовский классификатор с мультиномиальным распределением.

(c) Мультиномиальное распределение лучше показало себя на выборке с целыми неотрицательными значениям признаков.

Если понимать это выражение, как то, что мультиномиальное показало себя лучше на целых числах, чем на вещественных, то оно ложно. Если же понимать это, как "мультиномиальное распределение показало себя на целых числах лучше, чем все остальные распределения", тот оно истинно.


(d) На вещественных признаках лучше всего сработал нормальное распределение.

Верные утверждения: (d), ((c) в зависимости от понимания)

```python
from scipy.stats import uniform
from scipy.stats import norm
from scipy.optimize import minimize
```

Сгенерируем 500 точек, равномерно распределённых на отрезке $[-100; 100]$

```python
x = uniform.rvs(loc=-5, scale=10, size=500)
eps = norm.rvs(loc=0, scale=0.2**(0.5), size=500)
y = x * 0.5 + 1 + eps
```

```python
plt.figure(figsize=(15, 10))
plt.scatter(x, y, alpha=0.5, label="sample")
plt.grid()
plt.legend()
plt.show()
```

```python
def MSE(a):
    return ((a[0] * x + a[1] - y)**2).sum()/500.
```

```python
a = [1, 1]
coef = minimize(MSE, a)
```

```python
print("Полученное уравнение {:.2f}*x + {:.2f}".format(coef.x[0], coef.x[1]))
```

```python
grid = np.linspace(-10, 10, 100000)
plt.figure(figsize=(15, 10))
plt.scatter(x, y, alpha=0.5, label="sample")
plt.plot(grid, coef.x[0]*grid + coef.x[1], label="MSE", color='r')
plt.grid()
plt.legend()
plt.xlim(-7, 7)
plt.show()
```

```python
x_1 = uniform.rvs(loc=-5, scale=10, size=75)
eps_1 = norm.rvs(loc=0, scale=0.2**(0.5), size=75)
y_1 = eps_1 - 1
new_x = np.hstack((x, x_1))
new_y = np.hstack((y, y_1))
```

```python
plt.figure(figsize=(15, 10))
plt.scatter(new_x, new_y, alpha=0.5, label="sample")
plt.grid()
plt.legend()
plt.xlim(-10, 10)
plt.show()
```

```python
def MSE(a):
    return ((a[0] * new_x + a[1] - new_y)**2).sum()/575.
a = [1, 1]
coef_mse = minimize(MSE, a)
print("MSE: Полученное уравнение {:.2f}*x + {:.2f}".format(coef_mse.x[0], coef_mse.x[1]))
def MAE(a):
    return (abs(a[0] * new_x + a[1] - new_y)).sum()/575.
coef_mae = minimize(MAE, a)
print("MAE: Полученное уравнение {:.2f}*x + {:.2f}".format(coef_mae.x[0], coef_mae.x[1]))
```

```python
grid = np.linspace(-10, 10, 100000)
plt.figure(figsize=(15, 10))
plt.scatter(new_x, new_y, alpha=0.5, label="sample")
plt.plot(grid, coef_mse.x[0]*grid + coef_mse.x[1], label="MSE", color='r')
plt.plot(grid, coef_mae.x[0]*grid + coef_mae.x[1], label="MAE", color='g')
plt.grid()
plt.legend()
plt.xlim(-10, 10)
plt.show()
```

По графикам легко видеть, что модель, оптимизирующая MSE менее устойчива к выбросам, чем модель, оптимизирующая MAE. Это

# 4.1
### $P(x^{(k)} \mid y) = \frac{1}{\sqrt{2\pi \sigma^2}}e^{-\frac{(x^{(k)} - \mu_{yk})^2}{2\sigma^2}}$

### $\hat{y} = \arg\max_y P(y) \prod_{k=1}^{n} P(x^{(k)} \mid y) = \arg\max_y \prod_{k=1}^{n} \frac{1}{\sqrt{2\pi \sigma^2}}e^{-\frac{(x^{(k)} - \mu_{yk})^2}{2\sigma^2}} = \arg\max_y e^{\sum_{k=1}^{n}-\frac{(x^{(k)} - \mu_{yk})^2}{2\sigma^2}}$

Мы хотим максимизировать степень экспоненты так как степень экспоненты отрицательна, то при увеличении числителя, фукнция будет уменьшаться. Отсюда получаем, что максимум функции достигается, когда $x^{(k)} \rightarrow \mu_{yk}$

# 4.2

Подсчитаем $True Positives, False Positives, True Negatives, False Negatives$. Пусть у нас есть два класса объектов. И к 1 относится $n$ объектов, а к 0 относится $m$ объектов. Тогда $TP = n \cdot p, FP = n \cdot (1 - p), TN = m \cdot(1 - p), FN = m \cdot p$. Подсчитаем теперь $TPR = \frac{TP}{TP + FN} = \frac{n \cdot p}{n \cdot p + m \cdot p} = \frac{n}{n+m}, FPR = \frac{FP}{FP+TN} = \frac{n \cdot (1-p)}{n \cdot (1-p) + m\cdot (1-p)} = \frac{n}{n+m}$. Откуда получаем $FPR = TPR$, то есть $ROC$-кривая выглядит как биссектриса угла, а метрика $ROC-AUC$ равна площади прямоугольного теругольника со стороной 1, или 0.5.
